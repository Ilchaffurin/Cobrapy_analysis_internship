```python
mitocore_human=cobra.io.read_sbml_model("Models/12918_2017_500_MOESM4_ESM.xml")
```

```python
dict_results_voie_metabolique=dict()
dict_results_accoa=dict()
```

```python
output25=Utils_Cobrapy.define_boundary_and_run_model(model=mitocore_human,O2_upper_bound=7.3,glucose_upper_bound=(0.09+0.043)*0.11,
    FA_upper_bound=(0.09+0.043)*0.89,hydroxybutyrate_upper_bound=0.02791,Acetoacetate_upper_bound=0.05)
output65=Utils_Cobrapy.define_boundary_and_run_model(model=mitocore_human,O2_upper_bound=7.3,glucose_upper_bound=2.33*(0.09+0.043)*0.45807,
    FA_upper_bound=2.33*(0.09+0.043)*0.45807,hydroxybutyrate_upper_bound=0.02791,Acetoacetate_upper_bound=0.05)

output85=Utils_Cobrapy.define_boundary_and_run_model(model=mitocore_human,O2_upper_bound=7.3,glucose_upper_bound=3.41*(0.09+0.043)*0.75,
    FA_upper_bound=3.41*(0.09+0.043)*0.25,hydroxybutyrate_upper_bound=0.02791,Acetoacetate_upper_bound=0.05)
```

```python
dict_results_voie_metabolique["25"]=output25[1]
dict_results_voie_metabolique["65"]=output65[1]
dict_results_voie_metabolique["85"]=output85[1]
Utils_Cobrapy.plot_voie_metabolique_ou_accoa(dict_results_voie_metabolique,"plot_human_voie_metabolique.png")
```
