# Objectif trouver les réactions suivantes : 
le signe à coté des réactions correspond à l'impact de CAPN3 sur ces réactions (de manière primaire ou secondaires)
## Réactions impactées par PGC1alpha :
##### détoxifications ROS
superoxide dismutase 2 (SOD2)(+), glutathione peroxidase 1 (GPX1) (+)
##### mitochondrie
citrate synthase(+),CoA dehydrogènase(+), ATP synthase (+)
## réactions impliquées dans le métabolisme des lipides
adipose TAG lipase (ATGL) (+) (pas dans le modèle ?)
## hausse de l'activité du métabolisme glycolyique 
glycogen phosphorylase (production de glucose-16phosphate depuis glycogen)
lactate dehydrogenase,aldolase
## Informations supplémentaires nécessaires
P38 MAPK moins fonctionnelle chez CAPN3KO




## Gene non induit par l'activité physique chez C3KO
LPL,CD36,FABP3,UCP3,UCP4,DGAT,PLINs,ATGL

```python
List_R_Mito=list()
List_R_ROS=list()
List_R_Lipids=list()
List_R_Glycol=list()
List_R_Other=list()
for id in ["CSm","ACOAD10m","ACOAD1fm","ACOAD8m","ACOAD9m","ATPS4m"] : #réactions dans la mitochondrie
    List_R_Mito.append(MusModel.reactions.get_by_id(id))
for id in ["SPODMx","SPODMn","SPODMm","SPODMe","SPODM","O2Stx","O2Stn","O2Stm","O2St","GTHPi"]: #réactions en rapport avec le système ROS
    List_R_ROS.append(MusModel.reactions.get_by_id(id))
for id in ["LPS","LPS2","LPS2e","LPS3","LPS3e","LPS4e","PI45PLC","PI45PLCn","PI4PLCn","PIPLCn","PLA2","DGAT"]: #réactions en rapport avec le métabolisme des lipides
    List_R_Lipids.append(MusModel.reactions.get_by_id(id))
for id in ["GLDBRAN","GLGNS1","GLPASE1","GLPASE2","LALDD","LCADi","LCADi_D","LCADi_Dm","LDH_Lm"]: #réactions en rapport avec le métabolisme glucose
    List_R_Glycol.append(MusModel.reactions.get_by_id(id))
for id in ["Htm"] : #autres réactions : Htm = réaction permettant le transport d'un H+
    List_R_Other.append(MusModel.reactions.get_by_id(id))
for liste in [List_R_Mito,List_R_ROS,List_R_Lipids,List_R_Glycol,List_R_Other] :
    print("###")
    for reaction in liste :
        print(str(reaction.name)+" : "+str(reaction))
```
