```python
List_R_Mito=list()
List_R_ROS=list()
List_R_Lipids=list()
List_R_Glycol=list()
List_R_Other=list()
for id in ["CSm","ACOAD10m","ACOAD1fm","ACOAD8m","ACOAD9m","ATPS4m"] : #ractions dans la mitochondrie
    List_R_Mito.append(MusModel.reactions.get_by_id(id))
for id in ["SPODMx","SPODMn","SPODMm","SPODMe","SPODM","O2Stx","O2Stn","O2Stm","O2St","GTHPi"]: #ractions en rapport avec le systme ROS
    List_R_ROS.append(MusModel.reactions.get_by_id(id))
for id in ["LPS","LPS2","LPS2e","LPS3","LPS3e","LPS4e","PI45PLC","PI45PLCn","PI4PLCn","PIPLCn","PLA2","DGAT"]: #ractions en rapport avec le mtabolisme des lipides
    List_R_Lipids.append(MusModel.reactions.get_by_id(id))
for id in ["GLDBRAN","GLGNS1","GLPASE1","GLPASE2","LALDD","LCADi","LCADi_D","LCADi_Dm","LDH_Lm"]: #ractions en rapport avec le mtabolisme glucose
    List_R_Glycol.append(MusModel.reactions.get_by_id(id))
for id in ["Htm"] : #autres ractions : Htm = raction permettant le transport d'un H+
    List_R_Other.append(MusModel.reactions.get_by_id(id))
for liste in [List_R_Mito,List_R_ROS,List_R_Lipids,List_R_Glycol,List_R_Other] :
    print("###")
    for reaction in liste :
        print(str(reaction.name)+" : "+str(reaction))
```
