```python
param={"content-type":"application/json","target_taxon":"10090"} #parametre pour toutes les réquetes GET (10090=mus_musculus)
with open("Gene_List_Mitocore.csv",'w') as f :
    f.write("Gene_id\treactions\thomologue_MM\n")
    for x in Mitocore.genes :
        associated_ids = (i.id for i in x.reactions)
        homologue="NA"
        if x.id.startswith("ENSG") :
            url="https://rest.ensembl.org/homology/id/%s"%(x.id)
            r=requests.get(url=url,params=param)
            data=r.json()
            try : # si le json est complet (si il existe des homologues)
                if data["data"][0]["homologies"][0]["target"]["species"] == "mus_musculus" :
                    homologue="yes" # s'il existe un homologue Mus_musculus
                else :
                    homologue ="no" # s'il existe pas d'homologue Mus_musculus
            except IndexError :
                homologue="no" # s'il existe pas d'homologue 
        string="{}\t{}\t{}\n".format(x.id, "[" + ",".join(associated_ids) + "]",homologue)
        print(string)
        f.write(string)
f.close()
        
```

## Lecture de la base de données BRENDA
#### Création d'un dictionnaire contenant les informations nécessaires
le dictionnaire pourra etre amélioré selon les nécessités futures
cette étape nécessite la base de donnée BRENDA téléchargée depuis https://www.brenda-enzymes.org/download_brenda_without_registration.php

```python
with open("../Brenda/brenda_download.txt","r") as db :
    dict_brenda=dict()
    last_ec=None
    line=db.readline()
    while line!="":
        if line !="\n":
            if line[0:2]=="ID" : # Quand on arrive sur des un nouveau ECN
                if line.split()[0] not in dict_brenda.keys(): #on verifie que l'ecn n'est pas dans le dico
                    print(line)
                    dict_brenda[line.split()[1]]={"In_Mus_musculus":False,"Count_citation":0,}
                    last_ec=line.split()[1]
            if line[0:2]=="PR": #on regarde les lignes "PR"
                data=line.split()
                if len(data)>1 : #permet d'esquiver la ligne PROTEINS de chaque enzyme
                    if data[2]=="Mus" and data[3]=="musculus": # si la proteine existe chez mus_musculus
                        dict_brenda[last_ec]["In_Mus_musculus"]=True
                        if len(data)==5 : #on compte le nombre de citation supportant l'existance de la proteine chez mus musculus
                            citation_data=data[4].strip(">").strip("<").split(",")
                        elif len(data)==7:
                            citation_data=data[6].strip(">").strip("<").split(",")
                            
                        dict_brenda[last_ec]["Count_citation"]=len(citation_data)
        line=db.readline()
```

```python
def parse_ecn_from_mitocore(ecn):
    
    if type(ecn)==list:
        return ecn
    else :
        ecn_list=ecn.split()
        for item in ecn_list:
            if item =="and" or item=="or" or item=="+":
                ecn_list.remove(item)
        return ecn_list
    
print(parse_ecn_from_mitocore("2.7.1.1 or 2.7.1.2 and 2.7.1.3"))

def request_ecn_from_keggid(keggid):
    if keggid=="N/A":
        return []
    url="http://rest.kegg.jp/get/"
    r=requests.get(url=url+keggid)
    if r.status_code==200:
        for ligne in str(r.content).split("\\n"):
            if ligne.split()[0]=="ENZYME":
                return ligne.split()[1:]
        return []
    else : 
        print("warning : keggid not in kegg : "+keggid)
        return []
print(request_ecn_from_keggid("R01518"))
print(request_ecn_from_keggid("N/A"))
print(request_ecn_from_keggid("R02740"))
```

## Récupération des EC-number à partir des ID kegg des réactions

```python
for reaction in Mitocore.reactions :
    print(reaction.id)
    try :
        keggid=reaction.notes["KEGG id"].split()[0].strip(",")
    except KeyError :
        keggid="N/A"
    ecn=request_ecn_from_keggid(keggid)
    reaction.notes["EC Number"]=ecn
```
