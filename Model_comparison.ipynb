```python
Mitocore=cobra.io.read_sbml_model("Models/12918_2017_500_MOESM4_ESM.xml")
Recon1Mus=cobra.io.load_json_model("Models/iMM1415.json")
```

```python
#Exemple de comparaison a effectuer
reactionMitocore=Mitocore.reactions.get_by_id("G6PDH2r")
print(reactionMitocore.notes['EC Number'])
reactioniMM=Recon1Mus.reactions.get_by_id("G6PDH2r")
print(reactioniMM.annotation['ec-code'])
print(reactionMitocore.notes)
print(reactioniMM.id)
```

```python
#tri des ractions par sous-systme
iMM_reaction_by_subsystem=dict()
mitocore_reaction_by_subsystem=dict()

for r in Recon1Mus.reactions :
    subsystem=r.subsystem
    if subsystem not in iMM_reaction_by_subsystem.keys():
        iMM_reaction_by_subsystem[subsystem]=list()
    iMM_reaction_by_subsystem[subsystem].append(r)
for r in Mitocore.reactions :
    try :
        subsystem=r.notes['SUBSYSTEM']
    except KeyError :
        subsystem="None"
    if subsystem not in mitocore_reaction_by_subsystem.keys() :
        mitocore_reaction_by_subsystem[subsystem]=list()
    mitocore_reaction_by_subsystem[subsystem].append(r)
print(iMM_reaction_by_subsystem.keys())
print(mitocore_reaction_by_subsystem.keys())
```

```python
#algo naif de comparaison des EC-number
compteur_common=0
with open("iMM1415-Mitocore_in_common_reactions.csv","w") as f :
    f.write("imm_id\tmitocore_id\tsubsystem\tec-code\n")
    for r_iMM in Recon1Mus.reactions:
        try :
            immcode=r_iMM.annotation['ec-code']
        except KeyError :
            immcode=None
        for r_Mitocore in Mitocore.reactions :
            try : 
                mitocorecode=r_Mitocore.notes["EC Number"]
            except KeyError :
                mitocorecode=None
            if immcode != None :
                if mitocorecode in immcode :
                    compteur_common+=1
                    string=str(r_iMM.id)+"\t"+str(r_Mitocore.id)+"\t"+str(r_iMM.subsystem)+"\t"+str(mitocorecode)+"\n"
                    f.write(string)
print(compteur_common)
```

```python
data_iMM=dict()
with open("iMM1415-Mitocore_in_common_reactions.csv","r") as f :
    line=f.readline()
    line=f.readline()
    while line != "":
        liste=line.strip("\n").split("\t")
        if liste[0] not in data_iMM.keys():
            data_iMM[liste[0]]={"subsystem" : [],"Corresponding_reaction" : [] }
        data_iMM[liste[0]]["Corresponding_reaction"].append(liste[1])
        if liste[2] not in data_iMM[liste[0]]["subsystem"] :
            data_iMM[liste[0]]["subsystem"].append(liste[2])
        line=f.readline()
data_mitocore=dict()
with open("iMM1415-Mitocore_in_common_reactions.csv","r") as f :
    line=f.readline()
    line=f.readline()
    while line != "":
        liste=line.strip("\n").split("\t")
        if liste[1] not in data_mitocore.keys():
            data_mitocore[liste[1]]={"subsystem" : [],"Corresponding_reaction" : [] }
        data_mitocore[liste[1]]["Corresponding_reaction"].append(liste[0])
        if liste[2] not in data_mitocore[liste[1]]["subsystem"] :
            data_mitocore[liste[1]]["subsystem"].append(liste[2])
        line=f.readline()
print(len(data_mitocore))
for reaction in data_mitocore :
    in_model_reaction=Mitocore.reactions.get_by_id(reaction)
    print("raction :\t"+str(reaction))
    print("sous-systmes :\t"+str(data_mitocore[reaction]["subsystem"]))
    print("Raction correspondantes : \t"+str(data_mitocore[reaction]["Corresponding_reaction"]))
    print(str(in_model_reaction)+"\n")
print(len(data_iMM))
```

```python
# dtrminer quel pourcentage de chaque sous systme est en commun entre les deux modles
# dans mitocore
with open("mitocore_in_common_subsystem_data.csv","w") as f:
    f.write("subsystem;in_common;total;percentage\n")
    
    for subsystem in mitocore_reaction_by_subsystem :
        nb_reaction=len(mitocore_reaction_by_subsystem[subsystem])
        compteur=0
        for reaction in mitocore_reaction_by_subsystem[subsystem] :
            if reaction.id in data_mitocore.keys():
                compteur+=1
        
        string=str(subsystem)+";"+str(compteur)+";"+str(nb_reaction)+";"+str(compteur/nb_reaction*100)+"%\n"
        print(string)
        f.write(string)
    
```

```python
#dans iMM
with open("iMM_in_common_subsystem_data.csv","w") as f :
    f.write("subsystem;in_common;total;percentage\n")
    for subsystem in iMM_reaction_by_subsystem :
        nb_reaction=len(iMM_reaction_by_subsystem[subsystem])
        compteur=0
        for reaction in iMM_reaction_by_subsystem[subsystem] :
            if reaction.id in data_iMM.keys():
                compteur+=1
    
        string=str(subsystem)+";"+str(compteur)+";"+str(nb_reaction)+" ;"+str(compteur/nb_reaction*100)+"%\n"
        print(string)
        f.write(string)
        
        
```
